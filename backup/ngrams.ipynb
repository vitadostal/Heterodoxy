```python
import os.path as pth
dir = '/home/jovyan/work/'
document = []
document.append("")
pers_max = 76

for pers in range(1, pers_max + 1):
    name = dir + "/depos-stop/" + str("{:02d}".format(pers)) + ".txt"
    if (pth.isfile(name)):
        file = open(name, "rt")
        data = file.read()
        document.append(data)
```

```python
from nltk.util import bigrams, trigrams, ngrams   
from nltk import word_tokenize
from nltk import pos_tag
from nltk import FreqDist

ngram_count = 2
display_limit = 2
depos_max = 76

freq = []
freq.append("")

for i in range(1, depos_max + 1):
    print('CHAPTER ', i, ': ', document[i][:60])
    
    fdist = FreqDist()   
    for token in list(ngrams(word_tokenize(document[i].lower()), ngram_count)):
        fdist[token] += 1 
    freq.append(fdist)
    
    for item in (freq[i].most_common(30)):
        if (int(item[1]) >= display_limit):
            print(item[1], ': ', item[0])
    print()
```
