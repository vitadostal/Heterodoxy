```python
from nltk.util import bigrams, trigrams, ngrams   
from nltk import word_tokenize
from nltk import pos_tag
from nltk import FreqDist

display_limit = 1
depos_max = 76

freq = []
freq.append("")

for i in range(1, depos_max + 1):
    print('CHAPTER ', i, ': ', document[i][:60])
    
    postagged = pos_tag(word_tokenize(document[i].lower()))
    verbs = filter(lambda w: w[1][0] == 'V' and w[1][1] == 'B', postagged)
    
    fdist = FreqDist()
    for arr in list(verbs):
        token = arr[0] + ' / ' + arr[1]        
        fdist[token] += 1
    freq.append(fdist)

    for item in (freq[i].most_common(30)):
        if (int(item[1]) >= display_limit):
            print(item[1], ': ', item[0])    
    
    print()
```
